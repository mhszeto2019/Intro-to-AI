```python
# Import relevant packages
from __future__ import absolute_import, division, print_function, unicode_literals
import tensorflow as tf
from tensorflow.keras import layers
from pgmpy.models import BayesianModel
from pgmpy.factors.discrete import TabularCPD
import sys
```

```python
#PART A
#importing mnist fashion datasets
mnist= tf.keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Normalize input between 0 and 1
# Very important
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
#PART B
#Adding layers to the model
model = tf.keras.Sequential()
#Creating an ANN with 1 input Layer and at least two hidden layer with at least 10 nodes
model.add(layers.Flatten(input_shape=(28, 28)))         # input layer
model.add(layers.Dense(100, activation=tf.nn.relu))      # one hidden layer (acitvation by relu)
model.add(layers.Dense(100, activation=tf.nn.relu))      # one hidden layer
```

```python
#PART C
model.add(layers.Dense(10, activation=tf.nn.softmax))   # one output layer with 10 outputs
```

```python
#PARD D
# Compile the model with appropriate Loss function. metrics is something you can monitor (but model does not optimize metrc)
model.compile(optimizer=tf.optimizers.Adam(), 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
# Run the stochastic gradient descent for specified epochs
epochs = 5
model.fit(train_images, train_labels, epochs=epochs)

```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print('Test accuracy:', test_acc)
```
